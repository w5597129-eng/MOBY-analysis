```python
import pandas as pd
import glob
import os

# 처리 대상 파일들
input_files = [
    # "data/processed/1118_features.csv",
    "data/processed/1118_features_fluctuating_yellow.csv",
    "data/processed/1118_features_misalignment_yellow.csv",
]

def clean_and_save(input_path):
    # 1) 로드
    df = pd.read_csv(input_path)

    # 2) 컬럼 자동 탐색
    kurt_cols = [col for col in df.columns if col.endswith("SpectralKurtosis")]
    skew_cols = [col for col in df.columns if col.endswith("SpectralSkewness")]

    # 3) NaN 채우기
    df[kurt_cols] = df[kurt_cols].fillna(3.0)
    df[skew_cols] = df[skew_cols].fillna(0.0)

    # 4) 새 파일명 생성
    base, ext = os.path.splitext(input_path)
    output_path = base + "_cleaned" + ext

    # 5) 저장
    df.to_csv(output_path, index=False)
    print(f"✔ 저장 완료: {output_path}")

# 모든 파일 처리
for file in input_files:
    clean_and_save(file)
```
